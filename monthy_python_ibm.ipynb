```python
dice = 0
for i in range(4):
    dice += randint(1,6)
    print(dice)
    
```

```python
dices = []
for i in range(100):
    num = randint(1,6)
    dices.append(num)
print(dices)
```

```python
dices.count(1)
```

```python
resultats=[]
for i in range(1,7):
    resultats.append(dices.count(i))
    print(resultats)
```

```python
tot=0
for i in range(1,7):
    tot +=dices.count(i)
print(tot)
```

```python
dicetot = []
for j in range(10000):
    dicet = 0
    for i in range(4):
        dicet+=randint(1,6)
    dicetot.append(dicet)
print(dicetot)
```

```python
resultats=[]
for i in range(4,25):
    resultats.append(dicetot.count(i))
    print(resultats)
```

```python
n=[]
for i in range(4,25):
    n.append(i)
print(n)
fig, ax = plt.subplots()
ax.bar(n,resultats)
```
