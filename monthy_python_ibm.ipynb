```python
wins=0
wining_door =0
player_choice = 0
games = 10000
for i in range(games):
    wining_door = randint(0,2)
    player_choice = randint(0,2)
    
    if (player_choice == wining_door):
        open_door = (randint(1,2)+wining_door)%3
    else:
        open_door = 3-(player_choice+wining_door)
        
    player_door = 3 - (open_door+player_choice)
    
    if (player_door==wining_door):
            wins+=1
print(wins/games)
```
